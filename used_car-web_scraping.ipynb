```python
#To extract the Car details from the page with "Ahmedabad" Location

driver.get(city_links[0])
    
links1=[]
    
time.sleep(4)

#scrolling the cursor below in order to get number of details added on the page
for _ in range(7000):
    driver.execute_script("window.scrollBy(0,500)")

    
    
for i in driver.find_elements_by_xpath('//div[@class="gsc_col-xs-7  carsName"]//a'):
    try: 
        links1.append(i.get_attribute('href'))
    except NoSuchElementException:
        continue
```

```python
Car_details1 = {}
Car_details1['Model'] = []
Car_details1['Make_Year'] = []
Car_details1['Driven_Kilometers'] = []
Car_details1['Fuel'] = []
Car_details1['Transmission'] = []
Car_details1['Owner(s)'] = []
Car_details1['Mileage'] = []
Car_details1['Engine'] = []
#Car_details1['Seats']  = []
Car_details1['Price'] = []
```

```python
for j in links1:
    driver.get(j)
    time.sleep(3)
    
    #Model
    try:
        Model = driver.find_element_by_xpath('//div[@class="paddingBorder clearfix"]/div/h1')
        Car_details1['Model'].append(Model.text.replace('200','').replace('201','').replace('202','').replace('1','').replace('2','').replace('3','').replace('4','').replace('5','').replace('6','').replace('7','').replace('8','').replace('9','').replace('0',''))        
    except NoSuchElementException:
        try:
            Model = driver.find_element_by_xpath('//div[@id="rf01"]/div[1]/div/div/main/div[2]/div[1]/div/div[2]/div[2]/div/div/div/div[1]/div[1]/h1[2]')
            Car_details1['Model'].append(Model.text.replace('200','').replace('201','').replace('202','').replace('1','').replace('2','').replace('3','').replace('4','').replace('5','').replace('6','').replace('7','').replace('8','').replace('9','').replace('0',''))        
        except NoSuchElementException:
            Car_details1['Model'].append(np.nan)
    #Make_Year
    try:
        Make_Year = driver.find_element_by_xpath('//div[@id="rf01"]/div[1]/div/div/main/div/div/div[1]/div[3]/div/div/div/ul/li[1]/div/div[2]')
        Car_details1['Make_Year'].append(Make_Year.text)
    except NoSuchElementException:
        try:
            year = driver.find_element_by_xpath('//div[@id="widget-Overview"]/div/div/div/div[1]/ul/li[1]/div/div')
            Car_details1['Make_Year'].append(year.text)         
        except NoSuchElementException:
            Car_details1['Make_Year'].append(np.nan)
    #Driven_Kilometers
    try:
        Driven_Kilometers = driver.find_element_by_xpath('//div[@id="rf01"]/div[1]/div/div/main/div/div/div[1]/div[3]/div/div/div/ul/li[4]/div/div[2]')
        Car_details1['Driven_Kilometers'].append(Driven_Kilometers.text.replace(',','').replace(' Kms',''))
    except NoSuchElementException:
        try:
            km = driver.find_element_by_xpath('//div[@id="widget-Overview"]/div/div/div/div[1]/ul/li[3]/div/div')
            Car_details1['Driven_Kilometers'].append(km.text.replace(' kms','').replace(',',''))
        except NoSuchElementException:
            Car_details1['Driven_Kilometers'].append(np.nan)  
                #Price
    try:
        Price = driver.find_element_by_xpath('//div[@id="rf01"]/div[1]/div/div/main/div/div/div[2]/div[1]/div/div/div/div[1]/div[2]/div[1]/div[1]/span[2]')
        Car_details1['Price'].append(Price.text.replace(',',''))
    except NoSuchElementException:
        try:
            price = driver.find_element_by_xpath('//div[@id="rf01"]/div[1]/div/div/main/div[2]/div[1]/div/div[2]/div[2]/div/div/div/div[1]/span/span')
            Car_details1['Price'].append(price.text.replace(',','').replace(' Lakh','').replace('*',''))          
        except NoSuchElementException:
            Car_details1['Price'].append(np.nan)  
        
    #Fuel
    try:
        Fuel = driver.find_element_by_xpath('//div[@id="rf01"]/div[1]/div/div/main/div/div/div[1]/div[3]/div/div/div/ul/li[3]/div/div[2]')
        Car_details1['Fuel'].append(Fuel.text)
    except NoSuchElementException:
        try:
            fuel = driver.find_element_by_xpath('//div[@id="widget-Overview"]/div/div/div/div[1]/ul/li[5]/div/div')
            Car_details1['Fuel'].append(fuel.text)
        except NoSuchElementException:
            Car_details1['Fuel'].append(np.nan)  
          
    #Owner(s)
    try:
        Owner = driver.find_element_by_xpath('//div[@id="rf01"]/div[1]/div/div/main/div/div/div[1]/div[3]/div/div/div/ul/li[6]/div/div[2]')
        Car_details1['Owner(s)'].append(Owner.text.replace('st Owner','').replace('nd Owner','').replace('rd Owner','').replace('th Owner',''))
    except NoSuchElementException:
        try:
            owners = driver.find_element_by_xpath('//div[@id="widget-Overview"]/div/div/div/div[1]/ul/li[4]/div/div')
            Car_details1['Owner(s)'].append(owners.text.replace('First Owner','1').replace('Second Owner','2').replace('Third Owner','3').replace('Fourth Owner','4').replace('Fifth Owner',''))
        except NoSuchElementException:
            Car_details1['Owner(s)'].append(np.nan)
        
    #Transmission
    try:
        Transmission = driver.find_element_by_xpath('//div[@id="rf01"]/div[1]/div/div/main/div/div/div[1]/div[3]/div/div/div/ul/li[8]/div/div[2]')
        Car_details1['Transmission'].append(Transmission.text)
    except NoSuchElementException:
        try:
            transmission = driver.find_element_by_xpath('//div[@id="widget-Overview"]/div/div/div/div[1]/ul/li[6]/div/div')
            Car_details1['Transmission'].append(transmission.text)
        except NoSuchElementException:
            Car_details1['Transmission'].append(np.nan)  
    #Mileage
    try:
        Mileage = driver.find_element_by_xpath('//div[@id="specification-wdght"]/div/div/div/ul/li[1]/div/div[2]')
        Car_details1['Mileage'].append(Mileage.text.replace(' kmpl',''))
    except NoSuchElementException:
        try:
            Mileage = driver.find_element_by_xpath('//div[@id="topspec"]/div[1]/div/div/div[2]/div[1]/div/section/div/div/ul/li[1]/div[2]')
            Car_details1['Mileage'].append(Mileage.text.replace('kmpl',''))
        except NoSuchElementException:
            Car_details1['Mileage'].append(np.nan)
               #Engine
    try:
        Engine = driver.find_element_by_xpath('//div[@id="specification-wdght"]/div/div/div/ul/li[2]/div/div[2]')
        Car_details1['Engine'].append(Engine.text.replace(' CC',''))
    except NoSuchElementException:
        try:
            Engine = driver.find_element_by_xpath('//div[@id="topspec"]/div[1]/div/div/div[2]/div[1]/div/section/div/div/ul/li[2]/div[2]')
            Car_details1['Engine'].append(Engine.text.replace(' CC',''))
        except NoSuchElementException:
            Car_details1['Engine'].append(np.nan)
```
